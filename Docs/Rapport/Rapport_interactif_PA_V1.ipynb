# MLP

```python
#cas de test
X = np.array([[1, 0], [0, 1], [0, 0], [1, 1]])
Y = np.array([1, 1, -1, -1])
X
```

```python
model2 = keras.models.Sequential()
model2.add(keras.layers.Dense(2, activation=keras.activations.tanh, input_dim=2))
model2.add(keras.layers.Dense(3, activation=keras.activations.tanh))
model2.add(keras.layers.Dense(1, activation=keras.activations.tanh))
```

```python
model2.save('mlp_before_training.keras')
np.sign(model2.predict(X))
```

```python
model2.compile(loss=keras.losses.mean_squared_error,
              optimizer=keras.optimizers.SGD(lr=0.01))


model2.fit(X, Y, epochs=1000, batch_size=1)
```

```python
np.sign(model2.predict(X))
```
